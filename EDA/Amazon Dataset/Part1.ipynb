When regex=False, it means that the replacement will treat the string to be replaced ("₹" or ",") as plain text, not as a regex pattern. 

This is useful for simple replacements, like the ones in your code, where no complex pattern matching is required.

On the other hand, if you set regex=True, the method would interpret the replacement string as a regular expression, enabling more complex operations like searching for patterns rather than exact matches.

In your code, since regex=False is specified, you are simply removing the literal characters "₹" and "," without invoking any pattern-based matching. Does that make sense?


```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace("₹", "", regex=False)
df['discounted_price'] = df['discounted_price'].str.replace(",", "", regex=False)
df['discounted_price'] = df['discounted_price'].astype('float64')


df['actual_price'] = df['actual_price'].str.replace("₹", '', regex=False)
df['actual_price'] = df['actual_price'].str.replace(",", '', regex=False)
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace("%", "", regex=False)
df['discounted_price'] = df['discounted_price'].astype('float64')

df.head()
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
df['discount_percentage'] = df['discount_percentage'] / 100
df.head()
```

Q8: What is the correlation between discounted_price and rating?

```python
# Ensure the 'rating' column is numeric by replacing invalid values
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows with NaN values in 'rating' or 'discounted_price'
df_cleaned = df.dropna(subset=['rating', 'discounted_price'])

# Calculate the correlation
corr = df_cleaned["discounted_price"].corr(df_cleaned["rating"])
print(corr)
```
